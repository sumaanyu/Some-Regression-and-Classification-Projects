```python
# check for skewness
col_names = df.columns

fig, ax = plt.subplots(len(col_names), figsize=(16,12))

for i, col_val in enumerate(col_names):

    sns.distplot(df[col_val], hist=True, ax=ax[i])
    ax[i].set_xlabel(col_val, fontsize=12)
    ax[i].set_ylabel('Count', fontsize=12)

plt.subplots_adjust(bottom = 0.01, top = 2, hspace = 0.5 )
plt.show()
```

```python
#relationship between two variables
sns.pairplot(df);
```

```python
#outlier detection 
col_names = df.columns

fig, ax = plt.subplots(len(col_names), figsize=(5,40))

for i, col_val in enumerate(col_names):

    sns.boxplot(y=df[col_val], ax=ax[i])
    ax[i].set_xlabel(col_val, fontsize=8)
```

```python
%matplotlib inline
from pandas.plotting import scatter_matrix
labs = df.columns
axarr = scatter_matrix(df[labs],figsize=(20,20),hist_kwds={'bins':30,'color':'black'}) # create scatter plots and histograms

# the plots in the upper-right corner are repetitive, so replace this with correlations for quicker analysis
for i,axlist in enumerate(axarr):
    for j,ax in enumerate(axlist):
        if i<j:
            ax.cla() # clears an Axes object
            ax.text(x=0.8*ax.get_xlim()[0]+0.2*ax.get_xlim()[1],y=0.6*ax.get_ylim()[0]+0.4*ax.get_ylim()[1],
                    s=np.round(np.corrcoef(df[labs[i]],df[labs[j]])[0,1],2),fontsize=20) # adds text to an Axes object
```

```python
X = df.iloc[:, :-1 ]
y = df.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import ElasticNetCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#EDA FIN!
```

```python
from sklearn.linear_model import ElasticNet
```

```python
alphas = 10**np.linspace(10,-2,100)*0.5
ElasticCV = ElasticNetCV(alphas = alphas, normalize = True)
ElasticCV.fit(X_train, y_train)
ElasticCV.alpha_
```

```python
ElasticCV = ElasticNetCV(alphas = alphas, normalize = True)
ElasticCV.fit(X, y)
```

```python
ElasticCV.coef_
```

```python
ElasticCV.intercept_
```

```python
model = ElasticNet(alpha = ElasticCV.alpha_, normalize = True)
model
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.score(X_test,y_test)*100
```

```python
model.coef_.reshape((len(model.coef_),1))  
```

```python
coefficients_array = np.append(ElasticCV.coef_, ElasticCV.intercept_)
indices = np.append(X.columns,"intercept")
coefficients = pd.DataFrame(coefficients_array,index = indices).rename(columns = {0: "coefficients"})
coefficients
```
